```python
# 观察缺失值
df1.isnull().sum()
```

```python
#缺失值处理

# total_rev_hi_lim：总周转高信用/信用额度。以mean填充
# tot_cur_bal：所有帐户的当前总余额。 以0填充
# tot_coll_amt：欠款总额。以0填充
# revol_util：循环利用率，或借款人相对于所有可用循环信贷使用的信贷额度。以mean填充
# collections_12_mths_ex_med:除医疗账单外12个月的欠款数量。以0填充
# inq_last_6mths：过去6个月的征信查询数目（不包括汽车及按揭查询）。以0填充
# acc_now_delinq: 借款人现在欠款的账户数量。以0填充
# pub_rec: 贬损公共记录的数量。以0填充
# open_acc；借款人信用档案中的未结信用额度。以mean填充
# total_acc：借款人信用档案中当前信用额度的总数。以mean填充
# delinq_2yrs：过去两年借款人信用档案中逾期30天以上的拖欠次数。以0填充
# annual_inc：借款人在注册期间自行报告的年收入。以mean填充。
# emp_length：工作年限。以0填充

cols_fill_with_zero = [
    'tot_cur_bal', 'tot_coll_amt',
    'collections_12_mths_ex_med', 'acc_now_delinq',
    'pub_rec', 'delinq_2yrs','emp_length'
]
cols_fill_with_mean = ['total_rev_hi_lim', 'revol_util', 'annual_inc']

df2 = df1
for col in cols_fill_with_zero:
    df2[col] = df1[col].fillna(0)

for col in cols_fill_with_mean:
    mean_of_col = df1[col].mean()
    df2[col] = df1[col].fillna(mean_of_col)

print(df2.shape)
df2.head()
```

```python
df2 = df2.dropna(axis=0)  # 剔除剩余行里有空值的行

df2.shape
```

```python
# 将数据框写入CSV文件
df2.to_csv('loan_filtered.csv', index=False)
```
